```python
import arxiv
#from genderize import Genderize
import pandas as pd
import datetime
import gender_guesser.detector as gender
```

```python
search = arxiv.Search(
    query='physics'
)

results=[]

for result in search.results():
    results.append(result)
```

```python
results[0].published < datetime.datetime(2013, 9, 4, 10, 32, 56, tzinfo=datetime.timezone.utc)
```

```python
startdate = datetime.datetime(2012, 1, 1, 0, 0, 0, tzinfo=datetime.timezone.utc)
d = gender.Detector()
gender_results=[]
not_found=-1
for result in results:
    authors = result.authors
    firstnames = [str(author).split(' ')[0] for author in authors]
    for name in firstnames:
        if name.find('.') == not_found and result.published > startdate:
            gender_results.append(d.get_gender(name))
            print(name, d.get_gender(name))
```

```python
import pandas as pd
gender = pd.DataFrame( {'gender': gender_results})
gender[gender['gender']=='male'].count()
for groupname, data in gender.groupby('gender'):
    print(groupname, len(data))
```
